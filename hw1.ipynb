````python
# add your code here
# 题目上述算法是一个非常低效的排序算法
# 称为“傻瓜排序”（Stooge Sort）。时间复杂度平均和最坏情况下约为O(n^2.7095)
import math
def stooge_sort(arr):
    def sort(left, right):
        n = right - left + 1

        if n == 2 and arr[left] > arr[right]:
            arr[left], arr[right] = arr[right], arr[left]

        elif n >= 3:
            m = math.ceil(2 * n / 3)

            sort(left, left + m - 1)
            sort(right - m + 1, right)
            sort(left, left + m - 1)

    sort(0, len(arr) - 1)
    return arr

A = [5, 2, 9, 1, 3, 7]
print("本题目算法排序前：")
print(A)
print("排序后：")
print(stooge_sort(A))


# 快速排序算法时间复杂度为O(nlogn)
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[0]

    left = []
    right = []

    for x in arr[1:]:
        if x <= pivot:
            left.append(x)
        else:
            right.append(x)

    return quick_sort(left) + [pivot] + quick_sort(right)


A = [5, 2, 9, 1, 3, 7]
print("快速排序前：")
print(A)
print("快速排序后：")
print(quick_sort(A))
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

````python
# 第1、2小问 算法复杂度就是O（N）
# 算法分析 最大连续子数组和问题。
# profits[i] 表示第 i 天持有股票能够获得的收益，所以如果从某一天买入，到之后某一天卖出，总利润就是这一段连续天数收益的累加和。
# 例如：profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
# 如果选择第 5 天买入，第 8 天卖出，则收益为：5 + 2 + (-1) + 3 = 9
# 在数组 profits 中，找到一段连续区间，使得这段区间的和最大。

# 核心思想
# 如果前面的累积收益已经小于 0，那么继续加上前面的部分只会让总收益变小，所以应该从当前天重新开始计算。
# current_sum < 0 时，重新开始
# current_sum >= 0 时，继续累加

# 递推关系 设：current_sum 表示以第 i 天结尾的最大连续收益
# current_sum = max(profits[i], current_sum + profits[i])
# 全局最大利润为：max_profit = max(max_profit, current_sum)

# add your code here
def find_max_profit(profits):
    # 如果数组为空，说明没有任何股票收益数据
    # 此时最大利润为 0，买入和卖出日期都不存在，用 -1 表示
    if len(profits) == 0:
        return 0, -1, -1

    # current_sum 表示当前正在计算的连续收益和
    # 一开始从第 1 天开始，所以初始化为 profits[0]
    current_sum = profits[0]

    # max_profit 表示目前找到的最大利润
    # 一开始也只能选择第 1 天，所以初始化为 profits[0]
    max_profit = profits[0]

    # start 表示当前连续收益区间的起点下标
    start = 0

    # best_start 表示最大利润对应的买入日期下标
    best_start = 0

    # best_end 表示最大利润对应的卖出日期下标
    best_end = 0

    # 从第 2 天开始遍历，因为第 1 天已经用于初始化
    for i in range(1, len(profits)):

        # 如果之前累积的 current_sum 小于 0，
        # 说明之前那一段收益会拉低总收益，
        # 所以从当前第 i 天重新开始计算连续收益
        if current_sum < 0:
            current_sum = profits[i]
            start = i

        # 如果之前累积的 current_sum 大于等于 0，
        # 说明前面的收益对当前结果还有帮助，
        # 所以继续把第 i 天的收益加进去
        else:
            current_sum += profits[i]

        # 如果当前连续收益 current_sum 超过了历史最大利润 max_profit，
        # 就更新最大利润，并记录这一段区间的起点和终点
        if current_sum > max_profit:
            max_profit = current_sum
            best_start = start
            best_end = i

    # 因为数组下标从 0 开始，而题目中的日期一般从第 1 天开始，
    # 所以返回时要加 1
    return max_profit, best_start + 1, best_end + 1


# 测试数据 1
profits1 = [3, 2, 1, -7, 5, 2, -1, 3, -1]

max_p1, buy_day1, sell_day1 = find_max_profit(profits1)

print("测试数据 1：")
print("最大利润:", max_p1)
print("买入日期: 第", buy_day1, "天")
print("卖出日期: 第", sell_day1, "天")


# 测试数据 2
profits2 = [-2, 4, -1, 6, -3, 2, -5]

max_p2, buy_day2, sell_day2 = find_max_profit(profits2)

print("测试数据 2：")
print("最大利润:", max_p2)
print("买入日期: 第", buy_day2, "天")
print("卖出日期: 第", sell_day2, "天")

# your algorithm time complexity is:O(n)
# 空间复杂度 算法只使用了几个变量，O(1)
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:

````python
# 题目如下
# 给定一个包含不同整数元素的数组 A[1...n]，
# 如果某个元素比它左右两边的邻居都小，
# 即 A[x] < A[x - 1] 且 A[x] < A[x + 1]，
# 那么称 A[x] 为局部最小值。

# A[1] > A[2]，并且 A[n - 1] < A[n]
# 说明数组左边一开始是下降趋势，右边最后是上升趋势，
# 因此数组中一定存在至少一个局部最小值。

# 第1小问：O(log n) 找到一个局部最小值

# 算法分析：该问题可以使用二分查找的思想
# 每次取数组中间位置 mid：
# 1. 如果 A[mid] 同时小于左右邻居，说明 A[mid] 就是局部最小值，直接返回。
# 2. 如果 A[mid - 1] < A[mid]，说明左边方向存在下降趋势，那么左半部分一定存在局部最小值，所以继续在左半部分查找。
# 3. 否则，说明 A[mid + 1] < A[mid]，右边方向存在下降趋势，那么右半部分一定存在局部最小值，所以继续在右半部分查找。

# add your code here
def find_local_minimum(A):
    n = len(A)

    low = 0
    high = n - 1

    while low <= high:
        mid = (low + high) // 2

        # 如果 mid 是第一个元素，就把左邻居看成正无穷
        # 如果 mid 是最后一个元素，就把右邻居看成正无穷
        # 这样可以统一处理边界情况
        left_value = A[mid - 1] if mid > 0 else float('inf')
        right_value = A[mid + 1] if mid < n - 1 else float('inf')

        # 如果当前元素比左右邻居都小，说明找到了局部最小值
        if A[mid] < left_value and A[mid] < right_value:
            return A[mid]

        # 如果左边元素比当前元素小，
        # 说明左边方向更低，局部最小值一定在左边
        elif mid > 0 and A[mid - 1] < A[mid]:
            high = mid - 1

        # 否则说明右边方向更低，
        # 局部最小值一定在右边
        else:
            low = mid + 1

    return None

# 测试第一小问
arr = [9, 3, 7, 2, 1, 4, 5]
result = find_local_minimum(arr)

print("第一小问：")
print(f"找到的一个局部最小值为: {result}")

# your algorithm time complexity is:O(log n) 因为每次查找都会舍弃一半的区间
# 算法只使用了 low、high、mid 等变量，所以空间复杂度为 O(1)。



# 第2小问：找出所有局部最小值

# 算法分析：
# 如果要找出数组中的所有局部最小值，需要从左到右遍历整个数组，
# 对每一个位置 i 判断：A[i] 是否小于左边邻居，并且小于右边邻居。

# 对于边界位置：
# 如果 i 是第一个元素，则它没有左邻居，可以把左邻居看成正无穷。
# 如果 i 是最后一个元素，则它没有右邻居，可以把右邻居看成正无穷。
# 如果满足条件，就把 A[i] 加入答案列表。

# add your code here
def find_all_local_minimum(A):
    n = len(A)
    # 用来保存所有局部最小值
    local_minimums = []
    for i in range(n):
        # 处理左邻居
        left_value = A[i - 1] if i > 0 else float('inf')
        # 处理右邻居
        right_value = A[i + 1] if i < n - 1 else float('inf')
        # 判断当前元素是否比左右邻居都小
        if A[i] < left_value and A[i] < right_value:
            local_minimums.append(A[i])
    return local_minimums

# 测试第二小问
arr = [9, 3, 7, 2, 1, 4, 5]
all_result = find_all_local_minimum(arr)

print()
print("第二小问：")
print(f"所有局部最小值为: {all_result}")
# your algorithm time complexity is:O(n)。
# 如果不计算输出结果占用的空间，额外空间复杂度为 O(1)；
# 如果计算保存所有局部最小值的列表，则空间复杂度为 O(k)，
# 其中 k 是局部最小值的个数。
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

````python
# 二三问求助大模型

# 第一小问：先用归并排序排序，再输出前 k 个数
# 归并排序时间复杂度：O(nlogn)
# 取前 k 个元素时间复杂度：O(k)
# 总时间复杂度：O(nlogn + k)，因为 k <= n，所以可简化为 O(nlogn)
# 空间复杂度：O(n)

# add your code here
def merge_sort(arr):
    # 当数组长度为 0 或 1 时，数组本身已经有序
    if len(arr) <= 1:
        return arr
    # 将数组分成左右两部分
    mid = len(arr) // 2
    # 递归排序左半部分
    left = merge_sort(arr[:mid])
    # 递归排序右半部分
    right = merge_sort(arr[mid:])
    # 合并两个已经排好序的数组
    return merge(left, right)

def merge(left, right):
    result = []
    i = 0
    j = 0
    # 比较 left 和 right 中的元素，把较小的元素加入 result
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    # 如果 left 中还有剩余元素，继续加入 result
    while i < len(left):
        result.append(left[i])
        i += 1
    # 如果 right 中还有剩余元素，继续加入 result
    while j < len(right):
        result.append(right[j])
        j += 1
    return result

def find_k_smallest_by_sort(nums, k):
    # 如果 k 小于等于 0，返回空列表
    if k <= 0:
        return []
    sorted_nums = merge_sort(nums)
    # 返回排序后的前 k 个数
    return sorted_nums[:k]

# 测试
nums = [8, 3, 5, 1, 9, 2, 7]
k = 3

result = find_k_smallest_by_sort(nums, k)

print("原数组：", nums)
print("排序后的数组：", merge_sort(nums))
print("最小的", k, "个数为：", result)

# your algorithm time complexity is:O(n log n)
# your algorithm space complexity is:O(n)


# 第二小问

# 算法分析：
# 先建立一个空的最小堆，然后把数组中的元素一个一个插入堆中。
# 建堆完成后，连续执行 k 次 extract_min，得到最小的 k 个数。

# 时间复杂度：
# 逐个插入 n 个元素，每次插入 O(logn)，建堆总时间复杂度为 O(nlogn)
# Extract-min 执行 k 次，每次 O(logn)，时间复杂度为 O(klogn)
# 总时间复杂度为 O(nlogn + klogn)，可写为 O((n+k)logn)

# 空间复杂度：
# 堆中存储 n 个元素，空间复杂度为 O(n)


# add your code here
class MinHeap:
    def __init__(self):
        self.heap = []

    def insert(self, value):
        self.heap.append(value)
        self._sift_up(len(self.heap) - 1)

    def extract_min(self):
        if len(self.heap) == 0:
            return None

        min_value = self.heap[0]
        last_value = self.heap.pop()

        if len(self.heap) > 0:
            self.heap[0] = last_value
            self._sift_down(0)

        return min_value

    def _sift_up(self, index):
        while index > 0:
            parent = (index - 1) // 2

            if self.heap[parent] <= self.heap[index]:
                break

            self.heap[parent], self.heap[index] = self.heap[index], self.heap[parent]
            index = parent

    def _sift_down(self, index):
        n = len(self.heap)

        while True:
            smallest = index
            left = 2 * index + 1
            right = 2 * index + 2

            if left < n and self.heap[left] < self.heap[smallest]:
                smallest = left

            if right < n and self.heap[right] < self.heap[smallest]:
                smallest = right

            if smallest == index:
                break

            self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
            index = smallest


def find_k_smallest_by_heap(nums, k):
    if k <= 0:
        return []

    heap = MinHeap()

    for num in nums:
        heap.insert(num)

    result = []

    for _ in range(min(k, len(nums))):
        result.append(heap.extract_min())

    return result



# 第三小问：数据量超过内存时，分块处理

# 算法思想：
# 假设计算机最多只能存储 1GB 数据。
# 如果全部数据无法一次性放入内存，就把数据分成若干块。
# 每次只读取一块数据，在这一块内部找出最小的 k 个数。
# 每一块都处理完后，把所有块得到的候选数再次合并，从中找出全局最小的 k 个数。

# 时间复杂度：
# 假设总共有 n 个数据，每个块内部用大小为 k 的最大堆维护最小的 k 个数。
# 每个元素最多进入堆或与堆顶比较一次，因此整体时间复杂度为 O(nlogk)。
# 最后候选数据再处理一次，规模远小于 n。
# 所以总时间复杂度可近似写为 O(nlogk)。

# 空间复杂度：
# 每次只需要保存一个数据块和大小为 k 的堆。
# 如果每块大小不超过 1GB，则满足内存限制。
# 额外堆空间为 O(k)。

class MaxHeap:
    def __init__(self):
        self.heap = []

    def insert(self, value):
        self.heap.append(value)
        self._sift_up(len(self.heap) - 1)

    def get_max(self):
        if len(self.heap) == 0:
            return None
        return self.heap[0]

    def replace_max(self, value):
        if len(self.heap) == 0:
            return

        self.heap[0] = value
        self._sift_down(0)

    def _sift_up(self, index):
        while index > 0:
            parent = (index - 1) // 2

            if self.heap[parent] >= self.heap[index]:
                break

            self.heap[parent], self.heap[index] = self.heap[index], self.heap[parent]
            index = parent

    def _sift_down(self, index):
        n = len(self.heap)

        while True:
            largest = index
            left = 2 * index + 1
            right = 2 * index + 2

            if left < n and self.heap[left] > self.heap[largest]:
                largest = left

            if right < n and self.heap[right] > self.heap[largest]:
                largest = right

            if largest == index:
                break

            self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index]
            index = largest


def find_k_smallest_in_block(nums, k):
    if k <= 0:
        return []

    heap = MaxHeap()

    for num in nums:
        if len(heap.heap) < k:
            heap.insert(num)
        else:
            if num < heap.get_max():
                heap.replace_max(num)

    # 堆中是当前块最小的 k 个数，但不是升序
    return merge_sort(heap.heap)


def find_k_smallest_large_data(blocks, k):
    candidates = []

    # 每个 block 表示一个能放入内存的数据块
    for block in blocks:
        block_k_smallest = find_k_smallest_in_block(block, k)
        candidates.extend(block_k_smallest)

    # 在所有候选数中继续找最小的 k 个
    return find_k_smallest_in_block(candidates, k)



# 测试
arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3

print("原数组：", arr)
print("第一小问结果：", find_k_smallest_by_sort(arr, k))
print("第二小问结果：", find_k_smallest_by_heap(arr, k))

blocks = [
    [5, 4, 3, 2],
    [6, 1, 88],
    [33, 22, 107]
]

print("第三小问结果：", find_k_smallest_large_data(blocks, k))
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

````python
# 碰撞规则：
# 1. 质量小的星球消失。
# 2. 质量大的星球保留。
# 3. 如果质量相同，向左运动的星球消失，向右运动的星球保留。

# 算法分析：
# 使用栈模拟星球运动。
# 从左到右遍历数组：
# 如果当前星球不会和栈顶星球相撞，就直接入栈。
# 如果栈顶是正数，当前星球是负数，说明会相撞。
# 此时比较二者的绝对值：
# 若栈顶质量小，则栈顶星球消失，继续和新的栈顶比较。
# 若栈顶质量大或相等，则当前向左运动的星球消失。

# 每个星球最多入栈一次、出栈一次。
# 所以时间复杂度为 O(n)，空间复杂度为 O(n)。

# add your code here
def asteroid_collision(A):
    stack = []

    for planet in A:
        alive = True

        # 只有 栈顶向右、当前星球向左 时，才会发生碰撞
        while stack and stack[-1] > 0 and planet < 0 and alive:
            # 栈顶星球质量小，栈顶消失
            if abs(stack[-1]) < abs(planet):
                stack.pop()

            # 栈顶星球质量大，当前星球消失
            elif abs(stack[-1]) > abs(planet):
                alive = False

            # 质量相同，题目规定向左运动的星球消失
            else:
                alive = False

        # 如果当前星球没有被撞掉，就加入栈中
        if alive:
            stack.append(planet)

    return stack

# 测试数据 1
A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
result1 = asteroid_collision(A1)

print("测试数据 1：")
print("最终星球情况:", result1)

# 测试数据 2
A2 = [23, -8, 9, -3, -7, 9, -23, 22]
result2 = asteroid_collision(A2)

print("测试数据 2：")
print("最终星球情况:", result2)

# your algorithm time complexity is:
# 时间复杂度：O(n)
# 空间复杂度：O(n)
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

````python
# 0-1 背包问题：
# 有 n 块大理石，每块大理石有重量和价值。
# 小推车最大承重为 capacity。
# 每块大理石只能选择装或者不装，不能拆分。
# 目标是在不超过承重的情况下，使总价值最大。

# 算法分析：
# 使用动态规划。
# dp[j] 表示当前容量为 j 时，能够获得的最大价值。

# 对于每一块大理石，有两种选择：
# 1. 不装这块大理石，价值不变。
# 2. 装这块大理石，价值变为 dp[j - weight] + value。

# 状态转移方程：
# dp[j] = max(dp[j], dp[j - weight] + value)

# 因为每块大理石只能使用一次，
# 所以容量 j 要从大到小遍历。

# add your code here
def knapsack(items, capacity):
    dp = [0] * (capacity + 1)

    for weight, value in items:
        # 从大到小遍历，防止同一块大理石被重复使用
        for j in range(capacity, weight - 1, -1):
            dp[j] = max(dp[j], dp[j - weight] + value)

    return dp[capacity]


# 测试数据
items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9

result = knapsack(items, capacity)

print("最大价值为:", result)

# your algorithm time complexity is:
# 时间复杂度：O(n * capacity)
# 空间复杂度：O(capacity)
````

e.g.:

1.

输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:

````python
# 给定一棵二叉搜索树，其中恰好有两个节点的值被错误交换。
# 要求在不改变树结构的情况下，恢复这棵二叉搜索树。


# 算法分析：
# 二叉搜索树的中序遍历结果应该是升序序列。
# 如果有两个节点被交换，那么中序遍历时会出现逆序情况。

# 用 prev 记录中序遍历的前一个节点。
# 如果发现 prev.val > current.val，说明这两个节点顺序不对。

# 第一次发现逆序时，记录 first = prev，second = current。
# 如果第二次发现逆序，只更新 second = current。
# 最后交换 first 和 second 的值即可。

# 时间复杂度：O(n)，其中 n 是二叉树节点数
# 空间复杂度：O(h)，其中 h 是二叉树高度，递归调用栈最多为 h 层


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def recover_tree(root):
    first = None
    second = None
    prev = None

    def inorder(node):
        nonlocal first, second, prev

        if node is None:
            return

        inorder(node.left)

        if prev is not None and prev.val > node.val:
            if first is None:
                first = prev
            second = node

        prev = node

        inorder(node.right)

    inorder(root)

    if first is not None and second is not None:
        first.val, second.val = second.val, first.val

    return root


def tree_to_list(root):
    # 按层序遍历输出，格式和题目中的 [1, 3, None, None, 2] 一致
    if root is None:
        return []

    result = []
    queue = [root]

    while queue:
        node = queue.pop(0)

        if node is None:
            result.append(None)
        else:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)

    # 删除末尾多余的 None
    while result and result[-1] is None:
        result.pop()

    return result


# 测试数据 1
# 输入：root = [1, 3, None, None, 2]
# 恢复后：[3, 1, None, None, 2]

root1 = TreeNode(1)
root1.left = TreeNode(3)
root1.left.right = TreeNode(2)

recover_tree(root1)

print("测试数据 1：")
print("恢复后的树:", tree_to_list(root1))


# 测试数据 2
# 输入：root = [3, 1, 4, None, None, 2]
# 恢复后：[2, 1, 4, None, None, 3]

root2 = TreeNode(3)
root2.left = TreeNode(1)
root2.right = TreeNode(4)
root2.right.left = TreeNode(2)

recover_tree(root2)

print("测试数据 2：")
print("恢复后的树:", tree_to_list(root2))
````

idea:

````python
# 给定 n 个非负整数，表示每个宽度为 1 的柱子的高度。
# 要求计算下雨之后，这些柱子之间最多可以接多少雨水。

# 算法分析：
# 对于位置 i，它能接的雨水量取决于：
# 左边最高柱子 left_max
# 右边最高柱子 right_max

# 当前位置能接的雨水量为：
# min(left_max, right_max) - height[i]

# 如果这个值大于 0，说明可以接水；
# 如果小于等于 0，说明当前位置不能接水。

# 使用双指针从两边向中间移动：
# left 指向左边，right 指向右边。
# 如果 height[left] < height[right]，
# 说明左边较低，当前能否接水由 left_max 决定。
# 否则，右边较低，当前能否接水由 right_max 决定。

# add your code here
def trap_rain_water(height):
    if len(height) == 0:
        return 0

    left = 0
    right = len(height) - 1

    left_max = 0
    right_max = 0

    water = 0

    while left < right:
        # 左边柱子更低，处理左边
        if height[left] < height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                water += left_max - height[left]

            left += 1

        # 右边柱子更低，处理右边
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                water += right_max - height[right]

            right -= 1

    return water


# 测试数据 1
height1 = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
result1 = trap_rain_water(height1)

print("测试数据 1：",height1)
print("能接的雨水量为:", result1)


# 测试数据 2
height2 = [4, 2, 0, 3, 2, 5]
result2 = trap_rain_water(height2)

print("测试数据 2：",height2)
print("能接的雨水量为:", result2)

# your algorithm time complexity is:O(n)
# your algorithm space complexity is:O(1)
````

````python
# 给定一棵二叉树，树中每个节点存放 0 到 9 之间的数字。
# 从根节点到叶子节点的一条路径可以组成一个数字。
# 要求计算所有根到叶子路径组成的数字之和。


# 算法分析：
# 使用深度优先搜索 DFS。
# 从根节点开始向下遍历。
# 每到一个节点，就把当前路径上的数字更新为：
# current = current * 10 + node.val

# 如果当前节点是叶子节点，
# 说明已经形成了一个完整数字，直接返回 current。

# 如果不是叶子节点，
# 就继续递归计算左子树和右子树的路径和。

# add your code here
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def sum_numbers(root):
    def dfs(node, current):
        # 空节点不产生数字
        if node is None:
            return 0

        # 更新当前路径表示的数字
        current = current * 10 + node.val

        # 如果是叶子节点，返回当前完整数字
        if node.left is None and node.right is None:
            return current

        # 否则继续计算左右子树
        return dfs(node.left, current) + dfs(node.right, current)

    return dfs(root, 0)


# 测试数据 1
# root = [1, 2, 3]
# 路径：1 -> 2 表示 12
# 路径：1 -> 3 表示 13
# 总和：12 + 13 = 25

root1 = TreeNode(1)
root1.left = TreeNode(2)
root1.right = TreeNode(3)

print("测试数据 1：")
print("所有路径数字之和为:", sum_numbers(root1))


# 测试数据 2
# root = [4, 9, 0, 5, 1]
# 路径：4 -> 9 -> 5 表示 495
# 路径：4 -> 9 -> 1 表示 491
# 路径：4 -> 0 表示 40
# 总和：495 + 491 + 40 = 1026

root2 = TreeNode(4)
root2.left = TreeNode(9)
root2.right = TreeNode(0)
root2.left.left = TreeNode(5)
root2.left.right = TreeNode(1)

print("测试数据 2：")
print("所有路径数字之和为:", sum_numbers(root2))

# your algorithm time complexity is:O(n)
# 空间复杂度：O(h)
````
